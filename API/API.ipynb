```python
import json
import pandas as pd

# Assuming response_btc is your API response
status_code = response_btc.status_code
data = response_btc.json()

# Print the raw data to understand its structure (debugging)
print(json.dumps(data, indent=2))

# Initialize lists to store data
dates = []
opens = []
highs = []
lows = []
closes = []
volumes = []

# Extract metadata
meta_data = data.get("Meta Data", {})
digital_currency_name = meta_data.get("Digital Currency Name", "Bitcoin")

# Extract time series data
time_series_key = "Time Series (Digital Currency Weekly)"
time_series = data.get(time_series_key, {})

for date_str, values in time_series.items():
    dates.append(date_str)
    opens.append(float(values.get("1. open", 0)))
    highs.append(float(values.get("2. high", 0)))
    lows.append(float(values.get("3. low", 0)))
    closes.append(float(values.get("4. close", 0)))
    volumes.append(float(values.get("5. volume", 0)))

# Create DataFrame
BitcoinDF = pd.DataFrame({
    "date": dates,
    "open": opens,
    "high": highs,
    "low": lows,
    "close": closes,
    "volume": volumes,
    "currency_name": digital_currency_name
})

# Convert date column to datetime and sort
BitcoinDF['date'] = pd.to_datetime(BitcoinDF['date'])
BitcoinDF = BitcoinDF.sort_values('date', ascending=False)

# Display the DataFrame
print(BitcoinDF.head())

# Optional: Save to CSV
BitcoinDF.to_csv('Bitcoin_Weekly_Prices.csv', index=False)
```

# Merge Both DataSet

```python
import pandas as pd

# Load datasets
bitcoin = pd.read_csv('Bitcoin_Weekly_Prices.csv')
ethereum = pd.read_csv('Ethereum_Weekly_Prices.csv')
Solana = pd.read_csv('Solana_Weekly_Data.csv')

# Add currency identifier
#bitcoin['currency'] = 'Bitcoin'
#ethereum['currency'] = 'Ethereum'

# Combine vertically
crypto = pd.concat([bitcoin, ethereum, Solana], ignore_index=True)

# Convert date to datetime and sort
crypto['date'] = pd.to_datetime(crypto['date'])
crypto = crypto.sort_values(['date'])

# Reset index
crypto = crypto.reset_index(drop=True)

# Save to CSV
crypto.to_csv('Combined_Crypto_Data.csv', index=False)

# Display result
print(crypto.head(10))
```

```python
import requests

url = 'import requests'

url = 'https://u.today/search/node?keys=Bitcoin'
response_f = requests.get(url)
response_f


```

```python
import requests
import bs4

response = requests.get("https://u.today/search/node?keys=Bitcoin")
# soup = bs4.BeautifulSoup(response.text, 'lxml') # with lxml package
document = bs4.BeautifulSoup(response.text, 'html.parser') 
document
```

```python
# book_cards = soup.select("ol.row li article.product_pod")
```

```python
# Option 1: Using CSS selectors (similar to your book_cards example)
Bitcoin_date = document.select("body.not_logged_in div.main-aside-container main.main-block div.search-result div.news__item div.news__item-head div.humble")
Bitcoin_date
```

```python
Bitcoin_dt = []
for Bitcoin in Bitcoin_date:
    
  Bitcoin_dt.append(Bitcoin.get_text())
  #print(Bitcoin.get_text())
Bitcoin_dt
```

```python
News_cards = document.select('div.news__item')
News_cards
```

```python
from datetime import datetime

Bitcoin_data = []
for card in News_cards:
    data = []
    date_bt = card.select_one('div.news__item-head div.humble')
    if date_bt != None:
        datestr = date_bt.get_text()
        data.append(datetime.strptime(datestr.split(" - ")[0], "%b %d, %Y").date())
    title_bt = card.select_one('div.news__item-title')
    if title_bt != None:
        data.append(title_bt.get_text())
    if len(data) == 2:
        Bitcoin_data.append(tuple(data))
    
Bitcoin_data
```

```python
Bitcoin_Dataf = pd.DataFrame(Bitcoin_data, columns = ['date', 'title'])
Bitcoin_Dataf
```

```python
# Optional: Save to CSV
Bitcoin_Dataf.to_csv('Bitcoin_Daily_News.csv', index=False)
```
