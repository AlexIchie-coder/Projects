```python
import json
import pandas as pd

# Assuming response_btc is your API response
status_code = response_btc.status_code
data = response_btc.json()

# Print the raw data to understand its structure (debugging)
print(json.dumps(data, indent=2))

# Initialize lists to store data
dates = []
opens = []
highs = []
lows = []
closes = []
volumes = []

# Extract metadata
meta_data = data.get("Meta Data", {})
digital_currency_name = meta_data.get("Digital Currency Name", "Bitcoin")

# Extract time series data
time_series_key = "Time Series (Digital Currency Weekly)"
time_series = data.get(time_series_key, {})

for date_str, values in time_series.items():
    dates.append(date_str)
    opens.append(float(values.get("1. open", 0)))
    highs.append(float(values.get("2. high", 0)))
    lows.append(float(values.get("3. low", 0)))
    closes.append(float(values.get("4. close", 0)))
    volumes.append(float(values.get("5. volume", 0)))

# Create DataFrame
BitcoinDF = pd.DataFrame({
    "date": dates,
    "open": opens,
    "high": highs,
    "low": lows,
    "close": closes,
    "volume": volumes,
    "currency_name": digital_currency_name
})

# Convert date column to datetime and sort
BitcoinDF['date'] = pd.to_datetime(BitcoinDF['date'])
BitcoinDF = BitcoinDF.sort_values('date', ascending=False)

# Display the DataFrame
print(BitcoinDF.head())

# Optional: Save to CSV
BitcoinDF.to_csv('Bitcoin_Weekly_Prices.csv', index=False)
```

```python
import pandas as pd

# Load datasets
bitcoin = pd.read_csv('Bitcoin_Weekly_Prices.csv')
ethereum = pd.read_csv('Ethereum_Weekly_Prices.csv')

# Add currency identifier
#bitcoin['currency'] = 'Bitcoin'
#ethereum['currency'] = 'Ethereum'

# Combine vertically
crypto = pd.concat([bitcoin, ethereum], ignore_index=True)

# Convert date to datetime and sort
crypto['date'] = pd.to_datetime(crypto['date'])
crypto = crypto.sort_values(['date'])

# Reset index
crypto = crypto.reset_index(drop=True)

# Save to CSV
crypto.to_csv('Combined_Crypto_Data.csv', index=False)

# Display result
print(crypto.head(10))
```

```python
import json
import pandas as pd
import requests

# Make the API request
response_game = requests.get('https://api.sampleapis.com/switch/games')
status_code = response_game.status_code
game_data = response_game.json()

# Print sample data structure
print(json.dumps(game_data[0], indent=2))

# Initialize lists to store data
names = []
genres = []
developers = []
publishers = []
na_release = []
eu_release = []
jp_release = []
au_release = []

# Extract game data
for game in game_data:
    names.append(game.get("name", ""))
    genres.append(", ".join(game.get("genre", [])))
    developers.append(", ".join(game.get("developers", [])))
    publishers.append(", ".join(game.get("publishers", [])))
    
    # Handle release dates
    release_dates = game.get("releaseDates", {})
    na_release.append(pd.to_datetime(release_dates.get("NorthAmerica"), errors='coerce'))
    eu_release.append(pd.to_datetime(release_dates.get("Europe"), errors='coerce'))
    jp_release.append(pd.to_datetime(release_dates.get("Japan"), errors='coerce'))
    au_release.append(pd.to_datetime(release_dates.get("Australia"), errors='coerce'))

# Create DataFrame
games_df = pd.DataFrame({
    "name": names,
    "genre": genres,
    "developer": developers,
    "publisher": publishers,
    "na_release_date": na_release,
    "eu_release_date": eu_release,
    "jp_release_date": jp_release,
    "au_release_date": au_release
})

# Convert release dates to datetime and sort
games_df = games_df.sort_values("na_release_date", ascending=False)

# Display the DataFrame
pd.set_option('display.max_columns', None)
print(games_df.head(10))

# Save to CSV
#games_df.to_csv('switch_games.csv', index=False)
```
