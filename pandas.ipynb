most & least frequent MPAA Rating in the dataset in terms of occurrences

```python
import pandas as pd

# assume df is your DataFrame with 'MPAA_Rating' column
mpaa_ratings_counts = df['MPAA Rating'].value_counts()
mpaa_ratings_counts = mpaa_ratings_counts.sort_values(ascending=False)
most_frequent_mpaa_rating = mpaa_ratings_counts.index[0]
least_frequent_mpaa_rating = mpaa_ratings_counts.index[-1]

print(most_frequent_mpaa_rating)
print(least_frequent_mpaa_rating)
```
