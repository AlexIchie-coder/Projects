Density and residual sugar are two crucial parameters in wine production and quality assessment.

Residual Sugar:
Measured in units of grams per liter (g/L) or grams per 100 milliliters (g/100mL)
Refers to the amount of unfermented sugars remaining in the wine after fermentation
Influenced by factors like grape variety, ripeness, fermentation conditions, and yeast strain
Higher residual sugar levels typically indicate a sweeter wine, while lower levels indicate a drier wine
Residual sugar can also affect the wine’s overall balance, with high levels potentially masking acidity and tannins


Density:

-Influenced by factors like grape variety, ripeness, fermentation conditions, and aging

-Higher density typically indicates a wine with more solids and potentially more tannins and acidity

-Lower density may indicate a wine with more water and potentially less tannins and acidity


Red Wine (Residual Sugar)

-Residual sugar levels in red wine are not always directly correlated with perceived sweetness, as other factors like acidity, tannins, and fruit flavors influence the overall taste experience.

-Dry red wines can still have some residual sugar, while sweet red wines may have very little.

Red Wine (Density)

-It’s worth noting that the density of wine can range from 980 to 1020 kg/m³, with red wines typically having a higher density due to their higher alcohol and tannin content.

```python


# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

sns.kdeplot(red_wine["residual sugar"], fill=True, ax=ax1)
ax1.set_xlabel("Residual Sugar")
ax1.set_ylabel("Density")
ax1.set_title("Red Wine - Density of Residual Sugar")

# Bottom subplot: Residual plot
sns.residplot(x="density", y="residual sugar", data=red_wine, ax=ax2)
ax2.set_xlabel("density")
ax2.set_ylabel("Residual")
ax2.set_title("Residual Plot")

plt.tight_layout()
plt.show()
```

White Wine (Residual Sugar)

-Residual sugar levels in white wine vary significantly across styles and regions.

-Dry white wines typically have RS levels below 2 g/L, while off-dry and sweet wines have higher levels.


White Wine (Density)

-Density has a moderate impact on white wine quality. Higher density wines tend to have a richer mouthfeel and more full-bodied character, while lower density wines may appear lighter and more crisp.

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

sns.kdeplot(white_wine["residual sugar"], fill=True, ax=ax1)
ax1.set_xlabel("Residual Sugar")
ax1.set_ylabel("Density")
ax1.set_title("White Wine - Density of Residual Sugar")

# Bottom subplot: Residual plot
sns.residplot(x="density", y="residual sugar", data=white_wine, ax=ax2)
ax2.set_xlabel("density")
ax2.set_ylabel("Residual")
ax2.set_title("Residual Plot")

plt.tight_layout()
plt.show()
```

Free Sulfur Dioxide and pH in Red Wine: 

Free sulfur dioxide (FSO2) is a crucial component in wine making, particularly in red wine production, as it serves as a preservative to prevent oxidation and microbial spoilage. The pH of red wine typically ranges from 3.0 to 4.0, and the free SO2 concentration is dependent on this pH level. At lower pH levels, the molecular form of SO2 is more prevalent, which is more effective in protecting the wine.

```python

sns.scatterplot(x='pH', y='free sulfur dioxide', data=red_wine)
plt.xlabel('pH')
plt.ylabel('Free Sulfur Dioxide (mg/L)')
plt.title('Red Wine - pH vs. FSO2 in Wine')
plt.show()
```

Optimal Free Sulfur Dioxide Concentration: 
The optimal concentration of FSO2 in white wine can vary based on the wine’s pH. Generally, for white wines with a pH between 3.00 and 3.20, a free SO2 concentration of 0 to 20 mg/L is recommended. However, this can vary depending on factors such as the wine’s intended aging period and its susceptibility to oxidation and microbial spoilage.

```python

sns.scatterplot(x='pH', y='free sulfur dioxide', data=white_wine)
plt.xlabel('pH')
plt.ylabel('Free Sulfur Dioxide (mg/L)')
plt.title('White Wine - pH vs.  in Wine')
plt.show()
```
